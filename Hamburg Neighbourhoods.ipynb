```python
import wikipedia

wikipedia.set_lang('de')
```

```python
urls_list =[]

#to search 
for n in dataset.index:
    print('Neighbourhood Is :' , n)
    query = "Hamburg-{}".format(n)
    url=wikipedia.page(query).url
    print('WIKIPEDIA LINK FOUND:',url)
    urls_list.append(url)

    
    
```

```python
from bs4 import BeautifulSoup
import requests
import re


import warnings
warnings.filterwarnings(action="once")
```

```python
lat=[]
longi=[]
for link in urls_list:
    print(link)
    html=requests.get(link, verify=True).text
    soup = BeautifulSoup(html, 'lxml')
    lati=soup.find('span', class_="latitude")
    coordinate= str(lati)    
    result = re.split('<span class="latitude">(.*)</span>',coordinate)
    #new_res=result[1].replace(u'\xa0', '')
    #deg, minutes, seconds,null=re.split('[°\′″]', new_res)
    #num=round(float(deg) + float(minutes)/60 + float(seconds)/(60*60), 6)
    #print(num)
    lat.append(round(float(result[1]),6)) 
    print(result[1])
    long=soup.find('span' , class_="longitude")
    coordinate= str(long)
    res = re.split('<span class="longitude">(.*)</span>',coordinate)
    num=round(float(res[1]),6)
    longi.append(num) 
    print(num)
```

```python
stop_point = []
for i in range(stop_df.shape[0]):
    p = [stop_df.stop_lat.iloc[i],stop_df.stop_lon.iloc[i]]
    stop_point.append(p)
stop_points = np.array(stop_point)
stop_points
```



Next, I used `Voronoi` Diagram. 



```python
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 15]
```

In this part, I find each bus stop `latitude` and `longitude` in its neighbourhood

```python
region_list = []

for p in stop_points: 
    #calculate the distance from each center
    dis_cent=np.sum((neigh_centeriod - p)**2, axis=1)
    #find the minimum distance from the center
    min_index = np.argmin(dis_cent)
    #finding the index of regions with min distance(ridge_points=indices points between wuch each voronio ridges lies)
    ridges = np.where(centeriod.ridge_points == min_index)[0]
    #next with take all columns with respect rows lieing in ridges.this forms our region
    boundary_region=np.array(centeriod.ridge_vertices)[ridges, :]
    #We have to stick all boundaries
    boundary_region_set = set(boundary_region.ravel())
    region = [x for x in centeriod.regions if set(x) == boundary_region_set][0]
    region_list.append(region)
print(len(region_list))
```

Creating region dataframe to have number of stops.

```python
region_df = pd.DataFrame(columns={'number_of_stops':np.nan})

for i in range(stop_df.shape[0]):
    region_df.at[i, 'number_of_stops'] = region_list[i]    
        
        
        
        
```

Scraping Data from `Foursquare API`

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        pass
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
LIMIT = 99 # limit of number of venues returned by Foursquare API
radius = 500 # define radius

Hamburg_venues = getNearbyVenues(names=result['Neighborhoods'],
                                   latitudes=result['latitude'],
                                   longitudes=result['longitude'])





```
